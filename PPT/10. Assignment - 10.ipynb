### Question 1
<aside>
Given an integerÂ `n`, returnÂ *`true`Â if it is a power of three. Otherwise, returnÂ `false`*.

An integerÂ `n`Â is a power of three, if there exists an integerÂ `x`Â such thatÂ `n == 3x`.

**Example 1:**

```
Input: n = 27
Output: true
Explanation: 27 = 33
```

**Example 2:**

```
Input: n = 0
Output: false
Explanation: There is no x where 3x = 0.

```

**Example 3:**

```
Input: n = -1
Output: false
Explanation: There is no x where 3x = (-1).
```
</aside>

````python
def isPowerOfThree(n):
    # If n is less than or equal to 0, return False
    if n <= 0:
        return False

    # If n is equal to 1, return True
    if n == 1:
        return True

    # If n is divisible by 3, recursively call the function with n divided by 3
    if n % 3 == 0:
        return isPowerOfThree(n // 3)

    # If none of the above conditions are met, return False
    return False

for i in range(3):
    n = int(input("Enter a number: "))
    print("Input n :{}".format(n))
    res = isPowerOfThree(n)
    print("Output : {}\n".format(res))
````

### Question - 2

You have a listÂ `arr`Â of all integers in the rangeÂ `[1, n]`Â sorted in a strictly increasing order. Apply the following algorithm onÂ `arr`:

- Starting from left to right, remove the first number and every other number afterward until you reach the end of the list.
- Repeat the previous step again, but this time from right to left, remove the rightmost number and every other number from the remaining numbers.
- Keep repeating the steps again, alternating left to right and right to left, until a single number remains.

Given the integerÂ `n`, returnÂ *the last number that remains in*Â `arr`.

**Example 1:**

```
Input: n = 9
Output: 6
Explanation:
arr = [1, 2,3, 4,5, 6,7, 8,9]
arr = [2,4, 6,8]
arr = [2, 6]
arr = [6]

```

**Example 2:**

```
Input: n = 1
Output: 1
```

````python
def lastRemaining(n, leftToRight=True):
    # Base case: if n is equal to 1, return 1
    if n == 1:
        return 1

    # Recursive case: if leftToRight is True
    if leftToRight:
        # Call the function recursively with n divided by 2 and leftToRight set to False
        return 2 * lastRemaining(n // 2, False)
    else:
        # If n is even
        if n % 2 == 0:
            # Call the function recursively with n divided by 2 and leftToRight set to True
            return 2 * lastRemaining(n // 2, True) - 1
        else:
            # Call the function recursively with n divided by 2 and leftToRight set to True
            return 2 * lastRemaining(n // 2, True)

for i in range(2):
    n = int(input("Enter a number: "))
    print("Input : n : {}".format(n))
    res = lastRemaining(n)
    print("Output : {}\n".format(res))
````

### Question - 3

Given a set represented as a string, write a recursive code to print all subsets of it. The subsets can be printed in any order.

**Example 1:**

Input :Â Â set = â€œabcâ€

Output :Â { â€œâ€, â€œaâ€, â€œbâ€, â€œcâ€, â€œabâ€, â€œacâ€, â€œbcâ€, â€œabcâ€}

**Example 2:**

Input :Â set = â€œabcdâ€

Output :Â { â€œâ€, â€œaâ€ ,â€abâ€ ,â€abcâ€ ,â€abcdâ€, â€œabdâ€ ,â€acâ€ ,â€acdâ€, â€œadâ€ ,â€bâ€, â€œbcâ€ ,â€bcdâ€ ,â€bdâ€ ,â€câ€ ,â€cdâ€ ,â€dâ€ }

````python
def print_subsets(string):
    # Initialize a list to store all the subsets
    subsets = [""]

    # Iterate over each character in the string
    for char in string:
        # Append the current character to all existing subsets
        for i in range(len(subsets)):
            subsets.append(subsets[i] + char)

    # Sort the subsets list
    subsets.sort(key=lambda x: (len(x), x))

    # Print all the subsets
    print(subsets)

for i in range(2):
    string = input("Enter a string: ")
    print("Input: -  \nset :{}".format(string))
    # Call the function with initial values
    print("Output : - ")
    print_subsets(string)
    print()
````

### Question - 5

We are given a string S, we need to find count of all contiguous substrings starting and ending with same character.

**Examples :**

Input  : S = "abcab"
Output : 7

There are 15 substrings of "abcab"
a, ab, abc, abca, abcab, b, bc, bca
bcab, c, ca, cab, a, ab, b
Out of the above substrings, there
are 7 substrings : a, abca, b, bcab,
c, a and b.

Input  : S = "aba"
Output : 4
The substrings are a, b, a and aba

````python
def count_contiguous_substrings(s):
    # Initialize a variable to store the count of contiguous substrings
    count = 0

    # Get the length of the string
    n = len(s)

    # Iterate over each character in the string
    for i in range(n):
        # Iterate over each character from i to the end of the string
        for j in range(i, n):
            # If the characters at positions i and j are equal
            if s[i] == s[j]:
                # Increment the count of contiguous substrings
                count += 1

    # Return the count of contiguous substrings
    return count

for i in range(2):
    n = str(input("Enter a string: "))
    print("Input n :{}".format(n))
    res = count_contiguous_substrings(n)
    print("Output : {}\n".format(res))
````

### Question - 6

TheÂ tower of Hanoi is a famous puzzle where we have three rods andÂ **N**Â disks. The objective of the puzzle is to move the entire stack to another rod. You are given the number of discsÂ **N**. Initially, these discs are in the rod 1. You need to print all the steps of discs movement so that all the discs reach the 3rdÂ rod. Also, you need to find the total moves.**Note:**Â The discs are arranged such that theÂ **top disc is numbered 1**Â and theÂ **bottom-most disc is numbered N**. Also, all the discs haveÂ **different sizes**Â and a bigger discÂ **cannot**Â be put on the top of a smaller disc. Refer the provided link to get a better clarity about the puzzle.

**Example 1:**

Input:
N = 2

Output:

move disk 1 from rod 1 to rod 2
move disk 2 from rod 1 to rod 3
move disk 1 from rod 2 to rod 3
3

Explanation:For N=2Â , steps will be

**Example 2:**

Input:
N = 3

Output:

move disk 1 from rod 1 to rod 3
move disk 2 from rod 1 to rod 2
move disk 1 from rod 3 to rod 2
move disk 3 from rod 1 to rod 3
move disk 1 from rod 2 to rod 1
move disk 2 from rod 2 to rod 3
move disk 1 from rod 1 to rod 3
7

Explanation:For N=3 , steps will be
as follows in the example and total
7 steps will be taken.

````python
def tower_of_hanoi(n, source, destination, auxiliary):
    # Base case: if n is equal to 1
    if n == 1:
        # Print the move
        print("move disk 1 from rod", source, "to rod", destination)
        # Return the number of moves (1 in this case)
        return 1

    # Initialize a variable to store the number of moves
    moves = 0

    # Move n-1 disks from source to auxiliary using destination as auxiliary
    moves += tower_of_hanoi(n-1, source, auxiliary, destination)

    # Move the nth disk from source to destination
    print("move disk", n, "from rod", source, "to rod", destination)
    moves += 1

    # Move n-1 disks from auxiliary to destination using source as auxiliary
    moves += tower_of_hanoi(n-1, auxiliary, destination, source)

    # Return the total number of moves
    return moves

for i in range(2):
    n = int(input("Enter a number: "))
    print("Input n :{}".format(n))
    res = tower_of_hanoi(n, 1, 3, 2)
    print("Total moves : {}\n".format(res))
````

Given a stringÂ **str**, the task is to print all the permutations ofÂ **str**. AÂ **permutation**Â is an arrangement of all or part of a set of objects, with regard to the order of the arrangement. For instance, the words â€˜batâ€™ and â€˜tabâ€™ represents two distinct permutation (or arrangements) of a similar three letter word.

**Examples:**

> Input:Â str = â€œcdâ€
> 
> 
> **Output:**Â cd dc
> 
> **Input:**Â str = â€œabbâ€
> 
> **Output:**Â abb abb bab bba bab bba
>

````python
def permutations(string, current_permutation, remaining_chars):
    # Base case: if there are no remaining characters
    if len(remaining_chars) == 0:
        # Print the current permutation
        print(current_permutation, end=" ")
        print()
    else:
        # Recursive case: iterate over each remaining character
        for i in range(len(remaining_chars)):
            # Call the function recursively with the current character added to the current permutation
            # and the current character removed from the remaining characters
            permutations(string, current_permutation + remaining_chars[i], remaining_chars[:i] + remaining_chars[i+1:])

for i in range(2):
    string = str(input("Enter a string: "))
    print("Input : - str : {} \n".format(string))
    print("Output : - ")
    permutations(string, "", string)
````

<aside>

Given a string, count total number of consonants in it. A consonant is an English alphabet character that is not vowel (a, e, i, o and u). Examples of constants are b, c, d, f, and g.

**Examples :**

Input : abc de

Output : 3
There are three consonants b, c and d.

Input : geeksforgeeks portal

Output : 12

</aside>

````python
def count_consonants(string):
    # Initialize a variable to store the count of consonants
    count = 0

    # Define a list of vowels
    vowels = ['a', 'e', 'i', 'o', 'u']

    # Convert the string to lowercase
    string = string.lower()

    # Iterate over each character in the string
    for char in string:
        # If the character is an alphabet and not a vowel
        if char.isalpha() and char not in vowels:
            # Increment the count of consonants
            count += 1

    # Return the count of consonants
    return count

for i in range(2):
    name = str(input("Enter a string: "))
    print("Input n :{} ".format(name))
    res = count_consonants(name)
    print("Output : {}\n".format(res))
````
