### Question - 1

Given a linked list of **N** nodes such that it may contain a loop.

A loop here means that the last node of the link list is connected to the node at position X(1-based index). If the link list does not have any loop, X=0.

Remove the loop from the linked list, if it is present, i.e. unlink the last node which is forming the loop.

**Example 1:**

```
Input:
N = 3
value[] = {1,3,4}
X = 2
Output:1
Explanation:The link list looks like
1 -> 3 -> 4
     ^    |
     |____|
A loop is present. If you remove it
successfully, the answer will be 1.

```

**Example 2:**
```
Input:
N = 4
value[] = {1,8,3,4}
X = 0
Output:1
Explanation:The Linked list does not
contains any loop.
```

**Example 3:**

```
Input:
N = 4
value[] = {1,2,3,4}
X = 1
Output:1
Explanation:The link list looks like
1 -> 2 -> 3 -> 4
^              |
|______________|
A loop is present.
If you remove it successfully,
the answer will be 1.
```


````python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def removeLoop(self, head):
        # Initialize fast and slow pointers
        fast = head.next
        slow = head
        
        # Move fast pointer two steps and slow pointer one step at a time
        while fast!=slow:
            
            # If fast pointer reaches end of list, there is no loop
            if fast is None or fast.next is None:
                return
            
            fast = fast.next.next
            slow = slow.next
        
        # Find size of loop
        size = 1
        fast = fast.next
  
        while fast!=slow:
            fast = fast.next
            size+=1

        # Move slow and fast pointers to head of list
        slow=head
        fast=head

        # Move fast pointer size-1 steps ahead of slow pointer
        for _ in range(size-1):
            fast=fast.next
 
        # Move both pointers one step at a time until they meet
        while fast.next != slow:
            fast=fast.next
            slow=slow.next

        # Break the loop by setting next of last node to None
        fast.next=None

# Take input from user for number of test cases
T = int(input("Enter number of test cases: "))

for t in range(T):
    # Take input from user for each test case
    N = int(input("Enter number of nodes: "))
    print(f"N : {N}")
    values = list(map(int, input("Enter node values separated by space: ").split()))
    print(f"value[] : {values}")
    X = int(input("Enter index of node to create loop (-1 for no loop): "))
    print(f"X : {X}")

    nodes = []
    for val in values:
        nodes.append(ListNode(val))

    for i in range(N-1):
        nodes[i].next = nodes[i+1]

    if X >= 0 and X < N:
        nodes[N-1].next = nodes[X]

    solution = Solution()

    solution.removeLoop(nodes[0])

    if nodes[N-1].next is None:
        output = 1
    else:
        output = 0

    print("Output: ", output)
    print()
````

### Question - 2
A number **N** is represented in Linked List such that each digit corresponds to a node in linked list. You need to add 1 to it.

**Example 1:**

```
Input:
LinkedList: 4->5->6
Output:457

```
**Example 2:**

```
Input:
LinkedList: 1->2->3
Output:124
```

````python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def add_one_to_linked_list(head):
    # Reverse the linked list
    prev = None
    curr = head
    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
    head = prev
    
    # Add one to the reversed linked list
    carry = 1
    curr = head
    while curr:
        sum = curr.data + carry
        carry = sum // 10
        curr.data = sum % 10
        prev = curr
        curr = curr.next

    # Reverse the linked list again to get original order
    prev = None
    curr = head
    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
    head = prev

    # Convert linked list to integer and return result
    result = ""
    curr = head
    while curr:
        result += str(curr.data)
        curr = curr.next

    return int(result)

def print_linked_list(head):
    # Print linked list in human-readable format
    curr = head
    while curr:
        print(curr.data, end='')
        if curr.next:
            print('->', end='')
        curr = curr.next
    print()

# Take input from user for number of test cases
T = int(input("Enter number of test cases: "))

for t in range(T):
    # Take input from user for each test case
    values = list(map(int, input("Enter node values separated by space: ").split()))
    print(f"LinkedList : {values}")
    head = Node(values[0])
    curr = head

    for val in values[1:]:
        node = Node(val)
        curr.next = node
        curr = node

    res = add_one_to_linked_list(head)

    print("Output : ", res)
    print()
````

### Question - 3

Given a Linked List of size N, where every node represents a sub-linked-list and contains two pointers:(i) a **next** pointer to the next node,(ii) a **bottom** pointer to a linked list where this node is head.Each of the sub-linked-list is in sorted order.Flatten the Link List such that all the nodes appear in a single level while maintaining the sorted order. **Note:** The flattened list will be printed using the bottom pointer instead of next pointer.

**Example 1:**

```
Input:
5 -> 10 -> 19 -> 28
|     |     |     |
7     20    22   35
|           |     |
8          50    40
|                 |
30               45
Output: 5-> 7-> 8- > 10 -> 19-> 20->
22-> 28-> 30-> 35-> 40-> 45-> 50.
Explanation:
The resultant linked lists has every
node in a single level.(Note:| represents the bottom pointer.)

```

**Example 2:**

```
Input:
5 -> 10 -> 19 -> 28
|          |
7          22
|          |
8          50
|
30
Output: 5->7->8->10->19->22->28->30->50
Explanation:
The resultant linked lists has every
node in a single level.

(Note:| represents the bottom pointer.)
```



````python
class Node:
    def __init__(self, data):
        self.data = data
        self.right = None
        self.down = None

def merge(a, b):
    # Function to merge two linked lists
    if a is None:
        return b
    if b is None:
        return a
    if a.data < b.data:
        result = a
        result.down = merge(a.down, b)
    else:
        result = b
        result.down = merge(a, b.down)
    return result

def flatten(root):
    # Base case: if there are no more nodes on the right
    if root is None or root.right is None:
        return root
    # Recur for the list on the right
    root.right = flatten(root.right)
    # Merge this list with the list on right side
    root = merge(root, root.right)
    return root

def printList(root):
    # Function to print the flattened linked list
    while root is not None:
        print(root.data, end=" ")
        root = root.down

def printListRight(root):
    # Function to print the linked list with right pointers
    while root is not None:
        print(root.data, end=" ")
        temp = root.down
        while temp is not None:
            print(temp.data, end=" ")
            temp = temp.down
        print()
        root = root.right

# Creating the linked list
head = Node(5)
head.right = Node(10)
head.right.right = Node(19)
head.right.right.right = Node(28)

head.down = Node(7)
head.down.down = Node(8)
head.down.down.down = Node(30)

head.right.down = Node(20)

head.right.right.down = Node(22)
head.right.right.down.down = Node(50)

head.right.right.right.down = Node(35)
head.right.right.right.down.down = Node(40)
head.right.right.right.down.down.down = Node(45)

# Printing the input linked list
print("Input:")
printListRight(head)

# Flattening the linked list
head = flatten(head)

# Printing the flattened linked list
print("Output:")
printList(head)
````

Construct a copy of the given list. The copy should consist of exactly **N** new nodes, where each new node has its value set to the value of its corresponding original node. Both the next and random pointer of the new nodes should point to new nodes in the copied list such that the pointers in the original list and copied list represent the same list state. None of the pointers in the new list should point to nodes in the original list.

For example, if there are two nodes **X** and **Y** in the original list, where **X.arb** **-->** **Y**, then for the corresponding two nodes **x** and **y** in the copied list, **x.arb --> y.**

````python
# Node class
class Node:
    def __init__(self, x):
        self.data = x
        self.next = None
        self.arb = None

# Function to create a copy of the linked list
def copyList(head):
    # Create a new node for each node in the original list and insert it between the current and next node
    curr = head
    while curr:
        new_node = Node(curr.data)
        new_node.next = curr.next
        curr.next = new_node
        curr = curr.next.next

    # Set the random pointers for the new nodes
    curr = head
    while curr:
        if curr.arb:
            curr.next.arb = curr.arb.next
        curr = curr.next.next

    # Separate the original and copied lists
    original = head
    copy = head.next
    temp = copy
    while original and copy:
        original.next = original.next.next if original.next else None
        copy.next = copy.next.next if copy.next else None
        original = original.next
        copy = copy.next

    # Return the head of the copied list
    return temp


# Function to create a linked list from an array of values
def createList(values):
    head = Node(values[0])
    curr = head
    for value in values[1:]:
        curr.next = Node(value)
        curr = curr.next
    return head

# Function to set the random pointers for a linked list using an array of pairs
def setRandomPointers(head, pairs):
    nodes = {}
    curr = head
    i = 1
    while curr:
        nodes[i] = curr
        curr = curr.next
        i += 1
    for a, b in pairs:
        nodes[a].arb = nodes[b]

# Test case 1
N1 = 4
M1 = 2
values1 = [1, 2, 3, 4]
pairs1 = [(1, 2), (2, 4)]

# Create the original linked list for test case 1
original_list1 = createList(values1)

# Set the random pointers for the original list for test case 1
setRandomPointers(original_list1, pairs1)

# Print the original list for test case 1
print('Original List for Test Case 1:')
printList(original_list1)

# Create a copy of the linked list for test case 1 using the copyList function
copied_list1 = copyList(original_list1)

# Print the copied list for test case 1
print('Copied List for Test Case 1: \n')
printList(copied_list1)
print()

# Test case 2
N2 = 4
M2 = 2
values2 = [1, 3, 5, 9]
pairs2 = [(1, 1), (3, 4)]

# Create the original linked list for test case 2
original_list2 = createList(values2)

# Set the random pointers for the original list for test case 2
setRandomPointers(original_list2, pairs2)

# Print the original list for test case 2
print('Original List for Test Case 2:')
printList(original_list2)
print()
# Create a copy of the linked list for test case 2 using the copyList function
copied_list2 = copyList(original_list2)

# Print the copied list for test case 2
print('Copied List for Test Case 2:')
printList(copied_list2)
````

### Question - 5

Given the `head` of a singly linked list, group all the nodes with odd indices together followed by the nodes with even indices, and return *the reordered list*.

The **first** node is considered **odd**, and the **second** node is **even**, and so on.

Note that the relative order inside both the even and odd groups should remain as it was in the input.

You must solve the problem in `O(1)` extra space complexity and `O(n)` time complexity.

**Example 1:**

```
Input: head = [1,2,3,4,5]
Output: [1,3,5,2,4]
```
**Example 2:**

```
Input: head = [2,1,3,5,6,4,7]
Output: [2,3,6,7,1,5,4]
```

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def oddEvenList(head):
    if not head or not head.next:
        return head

    oddHead = head
    evenHead = head.next
    oddTail = oddHead
    evenTail = evenHead

    current = evenHead.next
    index = 3 

    while current:
        if index % 2 == 1:  
            oddTail.next = current
            oddTail = oddTail.next
        else:  # Even index
            evenTail.next = current
            evenTail = evenTail.next

        current = current.next
        index += 1

    oddTail.next = evenHead
    evenTail.next = None

    return oddHead


head1 = ListNode(1)
head1.next = ListNode(2)
head1.next.next = ListNode(3)
head1.next.next.next = ListNode(4)
head1.next.next.next.next = ListNode(5)

# Print the input values for test case 1
print('Input values for Test Case 1:')
current1 = head1
while current1:
    print(current1.val, end=" ")
    current1 = current1.next
print(end="\n")

reordered1 = oddEvenList(head1)

# Print the output values for test case 1
print('Output values for Test Case 1:')
current1 = reordered1
while current1:
    print(current1.val, end=" ")
    current1 = current1.next
    
print(end="\n")

head2 = ListNode(2)
head2.next = ListNode(1)
head2.next.next = ListNode(3)
head2.next.next.next = ListNode(5)
head2.next.next.next.next = ListNode(6)
head2.next.next.next.next.next = ListNode(4)
head2.next.next.next.next.next.next = ListNode(7)

# Print the input values for test case 2
print('Input values for Test Case 2:')
current2 = head2
while current2:
    print(current2.val, end=" ")
    current2 = current2.next
print(end="\n")

reordered2 = oddEvenList(head2)

# Print the output values for test case 2
print('Output values for Test Case 2:')
current2 = reordered2
while current2:
    print(current2.val, end=" ")
    current2 = current2.next # corrected line 
````

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def leftShiftLinkedList(head, k):
    if not head or not head.next or k == 0:
        return head

    current = head
    for _ in range(k-1):
        if not current.next:
            break
        current = current.next

    if not current.next:
        return head

    new_head = current.next
    last_node = current

    while current.next:
        current = current.next

    current.next = head

    last_node.next = None

    return new_head


head1 = ListNode(2)
head1.next = ListNode(4)
head1.next.next = ListNode(7)
head1.next.next.next = ListNode(8)
head1.next.next.next.next = ListNode(9)

# Print the input values for test case 1
print('Input - 1:')
current1 = head1
while current1:
    print(current1.val, end=" ")
    current1 = current1.next
print(end="\n")

k1 = 3
shifted1 = leftShiftLinkedList(head1, k1)

# Print the output values for test case 1
print('Output - 1:')

current1 = shifted1
while current1:
    print(current1.val, end=" ")
    current1 = current1.next
    
print(end="\n")

print()
    
head2 = ListNode(1)
head2.next = ListNode(2)
head2.next.next = ListNode(3)
head2.next.next.next = ListNode(4)
head2.next.next.next.next = ListNode(5)
head2.next.next.next.next.next = ListNode(6)
head2.next.next.next.next.next.next = ListNode(7)
head2.next.next.next. next. next. next. next=ListNode(8)

# Print the input values for test case 2
print('Input - 2:')
current2 = head2
while current2:
    print(current2.val, end=" ")
    current2 = current2. next
print(end="\n")

k2 = 4
shifted2 = leftShiftLinkedList(head2, k2)

# Print the output values for test case 2
print('Output - 2:')
current2 = shifted2
while current2:
    print(current2.val, end=" ")
    current2=current2. next # corrected line 
````

### Question - 7

You are given the `head` of a linked list with `n` nodes.

For each node in the list, find the value of the **next greater node**. That is, for each node, find the value of the first node that is next to it and has a **strictly larger** value than it.

Return an integer array `answer` where `answer[i]` is the value of the next greater node of the `ith` node (**1-indexed**). If the `ith` node does not have a next greater node, set `answer[i] = 0`.

**Example 1:**

```
Input: head = [2,1,5]
Output: [5,5,0]

```
**Example 2:**

```
Input: head = [2,7,4,3,5]
Output: [7,0,5,5,0]
```

</aside>


````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def nextLargerNodes(head):
    lst = []
    while head:
        lst.append(head.val)
        head = head.next

    n = len(lst)
    stack = []
    result = [0] * n

    for i in range(n-1, -1, -1):
        while stack and lst[i] >= stack[-1]:
            stack.pop()

        if stack:
            result[i] = stack[-1]

        stack.append(lst[i])

    return result


head1 = ListNode(2)
head1.next = ListNode(1)
head1.next.next = ListNode(5)

# Print the input values for test case 1
print('Input - 1:')
current1 = head1
while current1:
    print(current1.val, end=" ")
    current1 = current1.next
print(end="\n")

res1 = nextLargerNodes(head1)

# Print the output values for test case 1
print('Output - 1:')
print(res1)

print()

head2 = ListNode(2)
head2.next = ListNode(7)
head2.next.next = ListNode(4)
head2.next.next.next = ListNode(3)
head2.next.next.next.next = ListNode(5)

# Print the input values for test case 2
print('Input - 2:')
current2 = head2
while current2:
    print(current2.val, end=" ")
    current2 = current2. next
print(end="\n")

res2 = nextLargerNodes(head2)

# Print the output values for test case 2
print('Output - 2:')
print(res2)
````

### Question - 8

<aside>

Given the `head` of a linked list, we repeatedly delete consecutive sequences of nodes that sum to `0` until there are no such sequences.

After doing so, return the head of the final linked list.  You may return any such answer.

(Note that in the examples below, all sequences are serializations of `ListNode` objects.)

**Example 1:**

```
Input: head = [1,2,-3,3,1]
Output: [3,1]
Note: The answer [1,2,1] would also be accepted.

```

**Example 2:**

```
Input: head = [1,2,3,-3,4]
Output: [1,2,4]

```

**Example 3:**

```
Input: head = [1,2,3,-3,-2]
Output: [1]
```

</aside>

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def removeZeroSumSublists(head):
    
    dummy = ListNode(0)
    dummy.next = head

    cumulative_sums = {}
    current_sum = 0
    node = dummy

    while node:
        current_sum += node.val

        if current_sum in cumulative_sums:
            prev = cumulative_sums[current_sum].next
            prev_sum = current_sum + prev.val
            while prev_sum != current_sum:
                del cumulative_sums[prev_sum]
                prev = prev.next
                prev_sum += prev.val
            cumulative_sums[current_sum].next = node.next
        else:
            cumulative_sums[current_sum] = node

        node = node.next

    return dummy.next


head1 = ListNode(1)
head1.next = ListNode(2)
head1.next.next = ListNode(-3)
head1.next.next.next = ListNode(3)
head1.next.next.next.next = ListNode(1)

# Print the input values for test case 1
print('Input - 1:')
current1 = head1
while current1:
    print(current1.val, end=" ")
    current1 = current1.next
print(end="\n")

res1 = removeZeroSumSublists(head1)

# Print the output values for test case 1
print('Output - 1:')
while res1:
    print(res1.val, end=" ")
    res1 = res1.next
    
print(end="\n")
    
head2 = ListNode(1)
head2.next = ListNode(2)
head2.next.next = ListNode(3)
head2.next.next.next = ListNode(-3)
head2.next.next.next.next = ListNode(4)

print()

# Print the input values for test case 2
print('Input - 2:')
current2 = head2
while current2:
    print(current2.val, end=" ")
    current2 = current2. next
print(end="\n")

res2 = removeZeroSumSublists(head2)

# Print the output values for test case 2
print('Output - 2:')
while res2:
    print(res2.val, end=" ")
    res2=res2. next # corrected line 
print(end="\n")
print()


head3 = ListNode(1)
head3.next = ListNode(2)
head3.next.next = ListNode(3)
head3.next.next.next = ListNode(-3)
head3.next.next.next. next=ListNode(-2)

# Print the input values for test case 3
print('Input - 3:')
current3=head3
while current3:
    print(current3.val,end=" ")
    current3=current3. next
print(end="\n")

res3=removeZeroSumSublists(head3)

# Print the output values for test case 3
print('Output - 3:')
while res3:
    print(res3.val,end=" ")
    res3=res3. next
````
