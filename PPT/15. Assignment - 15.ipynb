### Question 1

<aside>

Given an array **arr[ ]** of size **N** having elements, the task is to find the next greater element for each element of the array in order of their appearance in the array.Next greater element of an element in the array is the nearest element on the right which is greater than the current element.If there does not exist next greater of current element, then next greater element for current element is -1. For example, next greater of the last element is always -1.

**Example 1:**

```
Input:
N = 4, arr[] = [1 3 2 4]
Output:
3 4 4 -1
Explanation:
In the array, the next larger element
to 1 is 3 , 3 is 4 , 2 is 4 and for 4 ?
since it doesn't exist, it is -1.

```

**Example 2:**

```
Input:
N = 5, arr[] [6 8 0 1 3]
Output:
8 -1 1 3 -1
Explanation:
In the array, the next larger element to
6 is 8, for 8 there is no larger elements
hence it is -1, for 0 it is 1 , for 1 it
is 3 and then for 3 there is no larger
element on right and hence -1.
```

</aside>


````python
def Next_Greater(arr, N):
    # Initialize an empty stack
    stack = []
    # Initialize the result array with all elements set to -1
    res = [-1] * N

    # Iterate over the array in reverse order
    for i in range(N - 1, -1, -1):
        # Pop elements from the stack that are less than or equal to the current element
        while stack and stack[-1] <= arr[i]:
            stack.pop()

        # If the stack is not empty, the next greater element is the top of the stack
        if stack:
            res[i] = stack[-1]

        # Push the current element onto the stack
        stack.append(arr[i])

    # Return the result array
    return res


# Example 1
arr1 = [1, 3, 2, 4]
N1 = len(arr1)
print(f"Input : N : {N1} , arr : {arr1} ")
output1 = Next_Greater(arr1, N1)
print(f"Output : {output1} \n")

# Example 2
arr2 = [6, 8, 0, 1, 3]
N2 = len(arr2)
print(f"Input : N : {N2} , arr : {arr2} ")
output2 = Next_Greater(arr2, N2)
print(f"Output : {output2}")
````

### Question 2

<aside>

Given an array **a** of integers of length **n**, find the nearest smaller number for every element such that the smaller element is on left side.If no small element present on the left print -1.

**Example 1:**

```
Input: n = 3
a = {1, 6, 2}
Output: -1 1 1
Explaination: There is no number at the
left of 1. Smaller number than 6 and 2 is 1.
```

**Example 2:**

```
Input: n = 6
a = {1, 5, 0, 3, 4, 5}
Output: -1 1 -1 0 3 4
Explaination: Upto 3 it is easy to see
the smaller numbers. But for 4 the smaller
numbers are 1, 0 and 3. But among them 3
is closest. Similary for 5 it is 4.
```

</aside>


````python
def Nearest_Smaller(a, n):
    # Initialize an empty stack
    stack = []
    # Initialize the result array with all elements set to -1
    res = [-1] * n

    # Iterate over the array
    for i in range(n):
        # Pop elements from the stack that are greater than or equal to the current element
        while stack and stack[-1] >= a[i]:
            stack.pop()

        # If the stack is not empty, the nearest smaller element is the top of the stack
        if stack:
            res[i] = stack[-1]

        # Push the current element onto the stack
        stack.append(a[i])

    # Return the result array
    return res


# Example 1
arr1 = [1, 6, 2]
n1 = len(arr1)
print(f"Input : N : {n1} , arr : {arr1} ")
output1 = Nearest_Smaller(arr1, n1)
print(f"Output : {output1} \n")

# Example 2
arr2 = [1, 5, 0, 3, 4, 5]
n2 = len(arr2)
print(f"Input : N : {n2} , arr : {arr2} ")
output2 = Nearest_Smaller(arr2, n2)
print(f"Output : {output2}")
````

Implement a Stack using two queues **q1** and **q2**.

**Example 1:**

```
Input:
push(2)
push(3)
pop()
push(4)
pop()
Output:3 4
Explanation:
push(2) the stack will be {2}
push(3) the stack will be {2 3}
pop()   poped element will be 3 the
        stack will be {2}
push(4) the stack will be {2 4}
pop()   poped element will be 4

```

**Example 2:**

```
Input:
push(2)
pop()
pop()
push(3)
Output:2 -1
```

</aside>


````python
from queue import Queue

class Stack:
    def __init__(self):
        self.q1 = Queue()
        self.q2 = Queue()

    def push(self, x):
        # Move all elements from q1 to q2
        while not self.q1.empty():
            self.q2.put(self.q1.get())
        # Put the new element in q1
        self.q1.put(x)
        # Move all elements back to q1 from q2
        while not self.q2.empty():
            self.q1.put(self.q2.get())

    def pop(self):
        # If q1 is empty, return -1
        if self.q1.empty():
            return -1
        # Return the front element of q1
        return self.q1.get()

# Example 1
print("Example 1:")
s = Stack()
print("push(2)")
s.push(2)
print("push(3)")
s.push(3)
print("pop()")
print("Output :" , s.pop()) # 3
print("push(4)")
s.push(4)
print("pop()")
print("Output :", s.pop()) # 4

# Example 2
print("\nExample 2:")
s = Stack()
print("push(2)")
s.push(2)
print("pop()")
print("Output : ", s.pop()) # 2
print("pop()")
print("Output : ", s.pop()) # -1
print(s.push(3))
s.push(3)
````

### Question 4
<aside>

You are given a stack **St**. You have to reverse the stack using recursion.

**Example 1:**

```
Input:St = {3,2,1,7,6}
Output:{6,7,1,2,3}
```

**Example 2:**

```
Input:St = {4,3,9,6}
Output:{6,9,3,4}
```

</aside>


````python
def reverse_Stack(St):
    # Base case: if the stack has 0 or 1 element, return
    if len(St) <= 1:
        return

    # Pop the top element from the stack
    top = St.pop()
    # Recursively reverse the remaining stack
    reverse_Stack(St)
    # Insert the top element at the bottom of the reversed stack
    insert_At_Bottom(St, top)


def insert_At_Bottom(St, item):
    # Base case: if the stack is empty, append the item
    if len(St) == 0:
        St.append(item)
        return

    # Pop the top element from the stack
    top = St.pop()
    # Recursively insert the item at the bottom of the stack
    insert_At_Bottom(St, item)
    # Append the top element back to the stack
    St.append(top)


# Example 1
St1 = [3, 2, 1, 7, 6]
print("Input:", St1)
reverse_Stack(St1)
print("Output:", St1)

# Example 2
St2 = [4, 3, 9, 6]
print("\nInput:", St2)
reverse_Stack(St2)
print("Output:", St2)
````

### Question 5
<aside>

You are given a string **S**, the task is to reverse the string using stack.

**Example 1:**

```
Input: S="GeeksforGeeks"
Output: skeeGrofskeeG
```

</aside>


````python
def reverse_String(S):
    # Initialize an empty stack
    stack = []
    # Push each character of the string onto the stack
    for char in S:
        stack.append(char)

    # Initialize an empty string to store the reversed string
    reversed_string = ""
    # Pop each character from the stack and append it to the reversed string
    while stack:
        reversed_string += stack.pop()

    # Return the reversed string
    return reversed_string

# Example
S = "GeeksforGeeks"
print("Input:", S)
reversed_S = reverse_String(S)
print("Output:", reversed_S)
````

### Question 6

<aside>

Given string **S** representing a postfix expression, the task is to evaluate the expression and find the final value. Operators will only include the basic arithmetic operators like ***, /, + and -**.

**Example 1:**

```
Input: S = "231*+9-"
Output: -4
Explanation:
After solving the given expression,
we have -4 as result.

```

**Example 2:**

```
Input: S = "123+*8-"
Output: -3
Explanation:
After solving the given postfix
expression, we have -3 as result.
```

</aside>


````python
def evaluate_Postfix_Expression(S):
    # Initialize an empty stack
    stack = []

    # Iterate over each character in the postfix expression
    for char in S:
        # If the character is a digit, push it onto the stack
        if char.isdigit():
            stack.append(int(char))
        # If the character is an operator, pop two operands from the stack,
        # perform the operation, and push the result back onto the stack
        else:
            operand2 = stack.pop()
            operand1 = stack.pop()

            if char == '+':
                res = operand1 + operand2
            elif char == '-':
                res = operand1 - operand2
            elif char == '*':
                res = operand1 * operand2
            elif char == '/':
                res = operand1 / operand2

            stack.append(res)

    # The final result is the top element of the stack
    return stack.pop()


# Example 1
S1 = "231*+9-"
print("Input:", S1)
res1 = evaluate_Postfix_Expression(S1)
print("Output:", res1)

# Example 2
S2 = "123+*8-"
print("\nInput:", S2)
res2 = evaluate_Postfix_Expression(S2)
print("Output:", res2)
````

### Question 7
<aside>

Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

Implement the `MinStack` class:

- `MinStack()` initializes the stack object.
- `void push(int val)` pushes the element `val` onto the stack.
- `void pop()` removes the element on the top of the stack.
- `int top()` gets the top element of the stack.
- `int getMin()` retrieves the minimum element in the stack.

You must implement a solution with `O(1)` time complexity for each function.

**Example 1:**

```
Input
["MinStack","push","push","push","getMin","pop","top","getMin"]
[[],[-2],[0],[-3],[],[],[],[]]

Output
[null,null,null,null,-3,null,0,-2]

Explanation
MinStack minStack = new MinStack();
minStack.push(-2);
minStack.push(0);
minStack.push(-3);
minStack.getMin(); // return -3
minStack.pop();
minStack.top();    // return 0
minStack.getMin(); // return -2
```

</aside>


````python
class Min_Stack:
    def __init__(self):
        # Initialize two stacks: one for the main stack and one for the minimum stack
        self.stack = []
        self.min_stack = []

    def push(self, val):
        # Push the value onto the main stack
        self.stack.append(val)
        # If the minimum stack is empty or the value is less than or equal to the top of the minimum stack,
        # push the value onto the minimum stack
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self):
        # If the main stack is not empty, pop the top element
        if self.stack:
            popped = self.stack.pop()
            # If the popped element is equal to the top of the minimum stack, pop it from the minimum stack as well
            if popped == self.min_stack[-1]:
                self.min_stack.pop()

    def top(self):
        # Return the top element of the main stack
        if self.stack:
            return self.stack[-1]

    def getMin(self):
        # Return the top element of the minimum stack
        if self.min_stack:
            return self.min_stack[-1]


def minStack(operations, values):
    output = []
    minStack = Min_Stack()
    for i in range(len(operations)):
        operation = operations[i]
        value = values[i]
        if operation == "MinStack":
            output.append(None)
        elif operation == "push":
            minStack.push(value[0])
            output.append(None)
        elif operation == "pop":
            minStack.pop()
            output.append(None)
        elif operation == "top":
            output.append(minStack.top())
        elif operation == "getMin":
            output.append(minStack.getMin())
    return output


# Example
operations = ["MinStack", "push", "push", "push", "getMin", "pop", "top", "getMin"]
values = [[], [-2], [0], [-3], [], [], [], []]
output = minStack(operations, values)
print("Input:")
print(operations)
print(values)
print("Output:")
print(output)  # [null,null,null,null,-3,null,0,-2]
````

### Question 8
<aside>

Given `n` non-negative integers representing an elevation map where the width of each bar is `1`, compute how much water it can trap after raining.

**Example 1:**

!https://assets.leetcode.com/uploads/2018/10/22/rainwatertrap.png

```
Input: height = [0,1,0,2,1,0,1,3,2,1,2,1]
Output: 6
Explanation: The above elevation map (black section) is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. In this case, 6 units of rain water (blue section) are being trapped.

```

**Example 2:**

```
Input: height = [4,2,0,3,2,5]
Output: 9
```

</aside>

````python
def trap(h):
    # Initialize left and right pointers
    l = 0
    r = len(h) - 1
    # Initialize maximum left and right heights
    max_left = max_right = 0
    # Initialize the amount of water trapped
    water_trapped = 0

    # While the left pointer is less than the right pointer
    while l < r:
        # If the height at the left pointer is less than or equal to the height at the right pointer
        if h[l] <= h[r]:
            # Update the maximum left height
            max_left = max(max_left, h[l])
            # Add the difference between the maximum left height and the current height to the water trapped
            water_trapped += max_left - h[l]
            # Move the left pointer to the right
            l += 1
        # If the height at the left pointer is greater than the height at the right pointer
        else:
            # Update the maximum right height
            max_right = max(max_right, h[r])
            # Add the difference between the maximum right height and the current height to the water trapped
            water_trapped += max_right - h[r]
            # Move the right pointer to the left
            r -= 1

    # Return the total amount of water trapped
    return water_trapped


# Example 1
h1 = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
print("Input:", h1)
print("Output:", trap(h1))

# Example 2
h2 = [4, 2, 0, 3, 2, 5]
print("\nInput:", h2)
print("Output:", trap(h2))
````
