#### Question 1  

Given an array, for each element find the value of the nearest element to the right which is having a frequency greater than that of the current element. If there does not exist an answer for a position, then make the value ‘-1’.

Examples:

Input: a[] = [1, 1, 2, 3, 4, 2, 1]

Output : [-1, -1, 1, 2, 2, 1, -1]

Explanation:

Given array a[] = [1, 1, 2, 3, 4, 2, 1]
Frequency of each element is: 3, 3, 2, 1, 1, 2, 3

Lets calls Next Greater Frequency element as NGF

1. For element a[0] = 1 which has a frequency = 3,
   As it has frequency of 3 and no other next element
   has frequency more than 3 so  '-1'
   
2. For element a[1] = 1 it will be -1 same logic
   like a[0]
   
3. For element a[2] = 2 which has frequency = 2,
   NGF element is 1 at position = 6  with frequency
   of 3 > 2
   
4. For element a[3] = 3 which has frequency = 1,
   NGF element is 2 at position = 5 with frequency
   of 2 > 1
   
5. For element a[4] = 4 which has frequency = 1,
   NGF element is 2 at position = 5 with frequency
   of 2 > 1
   
6. For element a[5] = 2 which has frequency = 2,
   NGF element is 1 at position = 6 with frequency
   of 3 > 2
   
7. For element a[6] = 1 there is no element to its
   right, hence -1

Input : a[] = [1, 1, 1, 2, 2, 2, 2, 11, 3, 3]

Output : [2, 2, 2, -1, -1, -1, -1, 3, -1, -1]


```python
from collections import Counter

def next_greater_frequency(arr):
    # Count the frequency of each element in the array
    freq = Counter(arr)
    
    # Initialize the result array with -1
    result = [-1] * len(arr)
    
    # Initialize a stack to keep track of elements
    stack = []
    
    # Iterate over the array
    for i in range(len(arr)):
        # If the stack is not empty and the frequency of the current element is greater than the frequency of the element at the top of the stack
        while stack and freq[arr[i]] > freq[arr[stack[-1]]]:
            # Pop the top element from the stack and update its value in the result array
            result[stack.pop()] = arr[i]
        
        # Push the current index to the stack
        stack.append(i)
    
    return result

# Test
arr = [1, 1, 2, 3, 4, 2, 1]
print(f'Input: {arr}')
print(f'Output: {next_greater_frequency(arr)} \n')

arr = [1, 1, 1, 2, 2, 2, 2, 11, 3, 3]
print(f'Input: {arr}')
print(f'Output: {next_greater_frequency(arr)}')
```

#### Question 2 

Given a stack of integers, sort it in ascending order using another temporary stack.

Examples:

Input : [34, 3, 31, 98, 92, 23]

Output : [3, 23, 31, 34, 92, 98]

Input : [3, 5, 1, 4, 2, 8]

Output : [1, 2, 3, 4, 5, 8]


```python
def sort_stack(stack):
    # Initialize a temporary stack
    temp_stack = []
    
    # While the input stack is not empty
    while stack:
        # Pop the top element from the input stack
        temp = stack.pop()
        
        # While the temporary stack is not empty and the top element of the temporary stack is greater than temp
        while temp_stack and temp_stack[-1] > temp:
            # Pop the top element from the temporary stack and push it to the input stack
            stack.append(temp_stack.pop())
        
        # Push temp to the temporary stack
        temp_stack.append(temp)
    
    return temp_stack

# Test
stack = [34, 3, 31, 98, 92, 23]
print(f'Input: {stack}')
print(f'Output: {sort_stack(stack)} \n')

stack = [3, 5, 1, 4, 2, 8]
print(f'Input: {stack}')
print(f'Output: {sort_stack(stack)}')
```

#### Question 3 
Given a stack with push(), pop(), and empty() operations, The task is to delete the middle element ****of it without using any additional data structure.

Input  : Stack[] = [1, 2, 3, 4, 5]

Output : Stack[] = [1, 2, 4, 5]

Input  : Stack[] = [1, 2, 3, 4, 5, 6]

Output : Stack[] = [1, 2, 4, 5, 6]


```python
def delete_middle(stack, n, curr=0):
    # Base case: if stack is empty or curr is equal to n
    if not stack or curr == n:
        return
    
    # Remove the top element from the stack
    temp = stack.pop()
    
    # Recursively remove the middle element
    delete_middle(stack, n, curr + 1)
    
    # If curr is not equal to n/2, push the element back to the stack
    if curr != n // 2:
        stack.append(temp)

# Test
stack = [1, 2, 3, 4, 5]
print(f'Input: {stack}')
delete_middle(stack, len(stack))
print(f'Output: {stack}\n')

stack = [1, 2, 3, 4, 5, 6]
print(f'Input: {stack}')
delete_middle(stack, len(stack))
print(f'Output: {stack}')
```

#### Question 5 
Given a number , write a program to reverse this number using stack.

Examples:

Input : 365

Output : 563

Input : 6899

Output : 9986


```python
def reverse_number(n):
    # Convert the number to a string
    n_str = str(n)
    
    # Initialize a stack
    stack = []
    
    # Push all digits to the stack
    for digit in n_str:
        stack.append(digit)
    
    # Initialize the result
    result = ''
    
    # Pop all digits from the stack and append them to the result
    while stack:
        result += stack.pop()
    
    # Convert the result to an integer and return it
    return int(result)

# Test
n = 365
print(f'Input: {n}')
print(f'Output: {reverse_number(n)} \n')

n = 6899
print(f'Input: {n}')
print(f'Output: {reverse_number(n)}')
```

#### Question 6  
Given an integer k and a queue of integers, The task is to reverse the order of the first k elements of the queue, leaving the other elements in the same relative order.

Only following standard operations are allowed on queue.

enqueue(x) : Add an item x to rear of queue

dequeue() : Remove an item from front of queue

size() : Returns number of elements in queue.

front() : Finds front item. 

```python
from queue import Queue

def reverse_first_k(q, k):
    # Check if k is valid
    if k < 0 or k > q.qsize():
        return
    
    # Initialize a stack
    stack = []
    
    # Dequeue the first k elements from the queue and push them to the stack
    for i in range(k):
        stack.append(q.get())
    
    # Pop all elements from the stack and enqueue them to the queue
    while stack:
        q.put(stack.pop())
    
    # Dequeue the remaining elements from the queue and enqueue them to the queue
    for i in range(q.qsize() - k):
        q.put(q.get())

# Test
q = Queue()
arr = [10, 20, 30, 40, 50, 60, 70, 80, 90]
for i in arr:
    q.put(i)
k = 5
print(f'Input: {arr}, k = {k}')
reverse_first_k(q, k)
print(f'Output: {list(q.queue)} \n')

q = Queue()
arr = [10, 20, 30, 40, 50]
for i in arr:
    q.put(i)
k = 4
print(f'Input: {arr}, k = {k}')
reverse_first_k(q, k)
print(f'Output: {list(q.queue)}')
```

####  Question 7 
Given a sequence of n strings, the task is to check if any two similar words come together and then destroy each other then print the number of words left in the sequence after this pairwise destruction.

Examples:

Input : ab aa aa bcd ab

Output : 3

As aa, aa destroys each other so,
ab bcd ab is the new sequence.

Input :  tom jerry jerry tom

Output : 0

As first both jerry will destroy each other.
Then sequence will be tom, tom they will also destroy
each other. So, the final sequence doesn’t contain any
word.


```python
def count_words_left(arr):
    # Initialize a stack
    stack = []
    
    # Iterate over the array
    for word in arr:
        # If the stack is not empty and the top element of the stack is equal to the current word
        if stack and stack[-1] == word:
            # Pop the top element from the stack
            stack.pop()
        else:
            # Push the current word to the stack
            stack.append(word)
    
    # Return the number of words left in the stack
    return len(stack)

# Test
arr = ['ab', 'aa', 'aa', 'bcd', 'ab']
print(f'Input: {arr}')
print(f'Output: {count_words_left(arr)} \n')

arr = ['tom', 'jerry', 'jerry', 'tom']
print(f'Input: {arr}')
print(f'Output: {count_words_left(arr)}')
```

####  Question 8 
Given an array of integers, the task is to find the maximum absolute difference between the nearest left and the right smaller element of every element in the array.

Note: If there is no smaller element on right side or left side of any element then we take zero as the smaller element. For example for the leftmost element, the nearest smaller element on the left side is considered as 0. Similarly, for rightmost elements, the smaller element on the right side is considered as 0.

Examples:

Input : arr[] = {2, 1, 8}

Output : 1

Left smaller  LS[] {0, 0, 1}

Right smaller RS[] {1, 0, 0}

Maximum Diff of abs(LS[i] - RS[i]) = 1

Input  : arr[] = {2, 4, 8, 7, 7, 9, 3}

Output : 4

Left smaller   LS[] = {0, 2, 4, 4, 4, 7, 2}

Right smaller  RS[] = {0, 3, 7, 3, 3, 3, 0}

Maximum Diff of abs(LS[i] - RS[i]) = 7 - 3 = 4

Input : arr[] = {5, 1, 9, 2, 5, 1, 7}

Output : 1


```python
def find_max_diff(arr):
    n = len(arr)
    
    # Initialize left smaller and right smaller arrays
    left_smaller = [0] * n
    right_smaller = [0] * n
    
    # Initialize a stack
    stack = []
    
    # Find left smaller elements
    for i in range(n):
        # While the stack is not empty and the top element of the stack is greater than or equal to the current element
        while stack and stack[-1] >= arr[i]:
            # Pop the top element from the stack
            stack.pop()
        
        # If the stack is not empty, update the left smaller element
        if stack:
            left_smaller[i] = stack[-1]
        
        # Push the current element to the stack
        stack.append(arr[i])
    
    # Clear the stack
    stack.clear()
    
    # Find right smaller elements
    for i in range(n - 1, -1, -1):
        # While the stack is not empty and the top element of the stack is greater than or equal to the current element
        while stack and stack[-1] >= arr[i]:
            # Pop the top element from the stack
            stack.pop()
        
        # If the stack is not empty, update the right smaller element
        if stack:
            right_smaller[i] = stack[-1]
        
        # Push the current element to the stack
        stack.append(arr[i])
    
    # Find maximum absolute difference between left smaller and right smaller elements
    max_diff = 0
    for i in range(n):
        max_diff = max(max_diff, abs(left_smaller[i] - right_smaller[i]))
    
    return max_diff

# Test
arr = [2, 1, 8]
print(f'Input: {arr}')
print(f'Output: {find_max_diff(arr)}\n')

arr = [2, 4, 8, 7, 7, 9, 3]
print(f'Input: {arr}')
print(f'Output: {find_max_diff(arr)}\n')

arr = [5, 1, 9, 2, 5, 1, 7]
print(f'Input: {arr}')
print(f'Output: {find_max_diff(arr)}')
```
