### Question - 1
<aside>


Given preorder of a binary tree, calculate its **[depth(or height)](https://www.geeksforgeeks.org/write-a-c-program-to-find-the-maximum-depth-or-height-of-a-tree/)** [starting from depth 0]. The preorder is given as a string with two possible characters.

1. ‘l’ denotes the leaf
2. ‘n’ denotes internal node

The given tree can be seen as a full binary tree where every node has 0 or two children. The two children of a node can ‘n’ or ‘l’ or mix of both.

**Examples :**

Input  : nlnll
Output : 2
Explanation :

!https://media.geeksforgeeks.org/wp-content/uploads/btree1.png

Input  : nlnnlll
Output : 3

!https://media.geeksforgeeks.org/wp-content/uploads/dia2.png

</aside>

```python
def calculateDepth(preorder):
    depth = 0

    for char in preorder:
        if char == 'n':
            depth += 1

    return depth

# Example 1
preorder1 = "nlnll"
print(f"Input : \nPreorder : {preorder1}")
print("Output : ",calculateDepth(preorder1))  # Output: 2

# Example 2
preorder2 = "nlnnlll"
print(f"\nInput : \nPreorder : {preorder2}")
print("Output : ",calculateDepth(preorder2))  # Output: 3
```

### Question  - 2
<aside>

Given a Binary tree, the task is to print the **left view** of the Binary Tree. The left view of a Binary Tree is a set of leftmost nodes for every level.

**Examples:**

***Input:***

            4

          /   \

        5     2

             /   \

            3     1

           /  \

          6    7

***Output:** 4 5 3 6*

**Explanation:**

!https://media.geeksforgeeks.org/wp-content/cdn-uploads/left-view.png

***Input:***

                    1

                  /   \

                2       3

                 \

                   4

                     \

                        5

                           \

                             6

**Output:** 1 2 4 5 6

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None


def printLeftView(root):
    if not root:
        return

    queue = [root]
    left_view = []

    while queue:
        size = len(queue)

        for i in range(size):
            node = queue.pop(0)

            # The first node at each level is the leftmost node
            if i == 0:
                left_view.append(node.val)

            # Add the left and right children to the queue
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

    return left_view


# Create the binary tree
root = TreeNode(4)
root.left = TreeNode(5)
root.right = TreeNode(2)
root.right.left = TreeNode(3)
root.right.right = TreeNode(1)
root.right.left.left = TreeNode(6)
root.right.left.right = TreeNode(7)

# Print the left view of the binary tree
left_view = printLeftView(root)
print("Left View: ", left_view)
```

### Question - 3
<aside>

Given a Binary Tree, print the Right view of it.

The right view of a Binary Tree is a set of nodes visible when the tree is visited from the Right side.

**Examples:**

**Input:**

         1

      /     \

   2         3

/   \       /  \

4     5   6    7

             \

               8

**Output**: 

Right view of the tree is 1 3 7 8

**Input:**

         1

       /

    8

  /

7

**Output**: 

Right view of the tree is 1 8 7

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None


def printRightView(root):
    if not root:
        return

    queue = [root]
    right_view = []

    while queue:
        size = len(queue)

        for i in range(size):
            node = queue.pop(0)

            # The last node at each level is the rightmost node
            if i == size - 1:
                right_view.append(node.val)

            # Add the left and right children to the queue
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

    return right_view


# Create the binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)
root.right.right.right = TreeNode(8)

# Print the right view of the binary tree
right_view = printRightView(root)
print("Right View: ", right_view)
```

### Question - 4
<aside>

Given a Binary Tree, The task is to print the **bottom view** from left to right. A node **x** is there in output if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1.

**Examples:**

**Input:**

             20

           /     \

        8         22

    /      \         \

5         3        25

        /    \

   10       14

**Output:** 5, 10, 3, 14, 25.

**Input:**

             20

           /     \

        8         22

    /      \      /   \

 5         3    4     25

         /    \

     10       14

**Output:**

5 10 4 14 25.

**Explanation:**

If there are multiple bottom-most nodes for a horizontal distance from the root, then print the later one in the level traversal.

**3 and 4** are both the bottom-most nodes at a horizontal distance of 0, we need to print 4.

</aside>

```python
from collections import deque

class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None
        self.hd = 0


def printBottomView(root):
    if not root:
        return

    # Initialize a dictionary to store the bottom-most nodes at each horizontal distance
    bottom_view = {}

    queue = deque([(root, 0)])

    while queue:
        node, hd = queue.popleft()

        # Update the horizontal distance of the current node
        node.hd = hd

        # Update the bottom view dictionary with the current node for its horizontal distance
        bottom_view[hd] = node.val

        # Add the left and right children to the queue
        if node.left:
            queue.append((node.left, hd - 1))
        if node.right:
            queue.append((node.right, hd + 1))

    # Print the bottom view nodes in the order of their horizontal distance
    bottom_view_nodes = [bottom_view[hd] for hd in sorted(bottom_view)]
    return bottom_view_nodes


# Create the binary tree
root = TreeNode(20)
root.left = TreeNode(8)
root.right = TreeNode(22)
root.left.left = TreeNode(5)
root.left.right = TreeNode(3)
root.right.right = TreeNode(25)
root.left.right.left = TreeNode(10)
root.left.right.right = TreeNode(14)

# Print the bottom view of the binary tree
bottom_view = printBottomView(root)
print("Bottom View: ", bottom_view)
```
